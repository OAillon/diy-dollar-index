```python
import pandas as pd
import numpy as np
import yfinance as yf

pd.set_option('display.max_columns', None)
```

## Getting data from YFinance

```python
tickers = [
    "EURUSD=X",
    "GBPUSD=X",
    "USDJPY=X",
    "USDMXN=X",
    "USDCHF=X"
]

raw = yf.download(
    tickers,
    start="2020-01-01",
    interval="1d",
    auto_adjust=True,
    progress=False
)

raw.tail()
```

### Small Cleanup & Log Returns Calculation

```python
fx = raw["Close"]
fx.columns = [c.replace("=X", "") for c in fx.columns]

log_returns = fx.apply(lambda x: pd.Series(np.log(x / x.shift(1))))
log_returns.dropna(inplace=True)

# fx = fx.reset_index()
display(fx.tail())
display(log_returns.tail())
```

# Putting it into a function

```python
def load_fx_rates(tickers, start, interval="1d"):
    raw = yf.download(
        tickers,
        start=start,
        interval=interval,
        auto_adjust=True,
        progress=False
    )
    fx = raw["Close"]
    fx.columns = [c.replace("=X", "") for c in fx.columns]
    return fx
```

```python
fx_func = load_fx_rates(tickers=tickers, start="2020-01-01", interval="1d")
fx_func.tail()
```

```python
fx_func.to_csv('../data/raw/raw_fx_data.csv')
```
